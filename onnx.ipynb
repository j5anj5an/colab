```python
from optimum.onnxruntime import ORTModelForSequenceClassification

# 定义模型名称
model_name = "distilbert-base-uncased"

# 加载预训练的模型和 tokenizer
model = ORTModelForSequenceClassification.from_pretrained(model_name, from_transformers=True)

# 导出模型到 ONNX 格式
model.save_pretrained("onnx")
```

```python
import time
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import onnxruntime as ort

def measure_time_pytorch(model_name, inputs, num_iterations=10):
    start_time = time.time()
    model = AutoModelForSequenceClassification.from_pretrained(model_name)
    load_time = time.time() - start_time
    model.eval()

    total_inference_time = 0.0

    for _ in range(num_iterations):
        start_time = time.time()
        with torch.no_grad():
            outputs = model(**inputs)
        inference_time = time.time() - start_time
        total_inference_time += inference_time

    avg_inference_time = total_inference_time / num_iterations

    return load_time, avg_inference_time

def measure_time_onnx(model_path, inputs, num_iterations=10):
    onnx_options = ort.SessionOptions()
    onnx_options.intra_op_num_threads = 1
    onnx_options.inter_op_num_threads = 1
    onnx_options.execution_mode = ort.ExecutionMode.ORT_SEQUENTIAL
    onnx_options.graph_optimization_level = ort.GraphOptimizationLevel.ORT_DISABLE_ALL
    # # 启用日志记录
    # options.log_severity_level = 0  # 日志所有消息
    # options.log_verbosity_level = 1  # 更详细的日志

    start_time = time.time()
    ort_session = ort.InferenceSession(model_path, onnx_options)
    load_time = time.time() - start_time

    print("\nExecution mode:", ort_session.get_session_options().execution_mode)
    print("intra_op_num_threads:", ort_session.get_session_options().intra_op_num_threads)
    print("inter_op_num_threads:", ort_session.get_session_options().inter_op_num_threads)
    print("inter_op_num_threads:", ort_session.get_session_options().graph_optimization_level)
    print("Provider:", ort_session.get_providers())
    total_inference_time = 0.0

    for _ in range(num_iterations):
        # 每次循环创建新的 ONNX Session
        ort_session = ort.InferenceSession(model_path)

        input_names = [input.name for input in ort_session.get_inputs()]
        ort_inputs = {name: inputs[name] for name in input_names}

        start_time = time.time()
        ort_outs = ort_session.run(None, ort_inputs)
        inference_time = time.time() - start_time
        total_inference_time += inference_time

    avg_inference_time = total_inference_time / num_iterations

    return load_time, avg_inference_time
```

```python
model_name = "distilbert-base-uncased"
onnx_model_path = "onnx/model.onnx"
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
text = """
grocery and food products to retailers, convenience stores and restaurants, from wal - mart stores, inc. see note 2 to the interim consolidated financial statements for additional information regarding the mclane acquisition. results of mclane \u2019 s business operations are included in berkshire \u2019 s consolidated results beginning on that date. mclane \u2019 s revenues in 2004 totaled $ 6, 117 million for the third quarter and $ 17, 352 million for the first nine months. pre - tax earnings totaled $ 62 million for the third quarter and $ 176 million for the first nine months of 2004. on a proforma basis, mclane \u2019 s sales and pre - tax earnings for the first nine months of 2003 totaled approximately $ 16, 262 million and $ 175 million, respectively. the comparative year - to - date increases in sales reflect the addition of new customers since berkshire \u2019 s acquisition and growth in the food service business. in 2003, approximately 35 % of mclane \u2019 s annual revenues derived from sales to wal - mart stores, inc. mclane \u2019 s business is marked by high sales volume and very thin profit margins. 19 management \u2019 s discussion and analysis of financial condition and results of operations ( continued ) non - insurance businesses ( continued ) revenues from berkshire \u2019 s various home furnishing and jewelry retailers for the third quarter of 2004 increased $ 47 million ( 8. 6 % ) over the third quarter of 2003. for the first nine months, retail revenues in 2004 increased $ 270 million ( 17. 9 % ) over 2003. same - store sales increased 1. 7 % in the third quarter and 3. 2 % for the first nine months. pre - tax earnings of the retailing group for the third quarter of 2004 totaled $ 25 million, relatively unchanged from 2003. for the first nine months of 2004, pre - tax earnings of $ 81 million increased $ 5 million ( 6. 6 % ) over 2003. pre - tax earnings benefited from higher sales volume, but were partially offset by increased depreciation and salary and benefit expenses.
"""
text_len = 1024
inputs = tokenizer(text[:text_len], return_tensors="pt")
onnx_inputs = tokenizer(text[:text_len], return_tensors="np")
```

```python
num_iterations = 10
print (f'times: {num_iterations}, len: {text_len}')

# 测量 ONNX 模型时间
ort.set_default_logger_severity(1)
onnx_load_time, onnx_inference_time = measure_time_onnx(onnx_model_path, onnx_inputs, num_iterations=num_iterations)
print(f"ONNX Model Load Time: {onnx_load_time:.4f} seconds")
print(f"ONNX Model Inference Time: {onnx_inference_time:.4f} seconds")
```

```python
# 设置 PyTorch 线程数
torch.set_num_threads(3)
# torch.set_num_interop_threads(3)
print("\nIntra-op torch threads is:", torch.get_num_threads())
print("Inter-op torch threads is:", torch.get_num_interop_threads())

# 测量 Pytorch 模型时间
pytorch_load_time, pytorch_inference_time = measure_time_pytorch(model_name, inputs, num_iterations=num_iterations)
print(f"pytorch Model Load Time: {pytorch_load_time:.4f} seconds")
print(f"pytorch Model Inference Time: {pytorch_inference_time:.4f} seconds")
```
